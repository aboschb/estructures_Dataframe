Anem a descriure les variables del nostre dataset:

- Unnamed: columna sense nom, per tant no sabem a quins valors es refereix.
- Year: any del vol.

Ara crearem una columna nova per veure la velocitat mitjana de cada vol. 
Tenint en compte que la velocitat mitjana és igual al desplaçament dividit entre el temps, utilitzarem les variables
Distance i AirTime (que té en compte el temps de vol únicament i no el temps en pista). 

```python
# Calculem la nova columna basada en les dues columnes existents. 
# La variable AirTime està en minuts, per tant la passem a hores. Arrodonim a 2 decimals el resultat obingut.

df_reduit['AverageAircraftSpeed'] = (df_reduit['Distance'] / (df_reduit['AirTime'] / 60)).round(2)

# Mostrem els 10 primers registres del dataframe amb la nova columna (els resultats són en milles per hora)
df_reduit.head(10)
```

Ara crearem una columna que ens mostrarà si el vol ha patit retard o no, tant en la sortida com en l'arribada. 

```python
# Creem la nova columna anomenada Delays i l'inicialitzem amb una string buida (on guardarem els resultats de les condicions)
df_reduit['Delays'] = ''

# Utilitzem la funció .loc per accedir als elements de les columnes del DataFrame
# i apliquem les condicions als valors individuals de cada fila
df_reduit.loc[df_reduit['ArrDelay'] > 0, 'Delays'] = "Arr Delay"
df_reduit.loc[df_reduit['DepDelay'] > 0, 'Delays'] = "Dep Delay"
df_reduit.loc[(df_reduit['ArrDelay'] > 0) & (df_reduit['DepDelay'] > 0), 'Delays'] = "Arr and Dep Delay"
df_reduit.loc[(df_reduit['ArrDelay'] <= 0) & (df_reduit['DepDelay'] <= 0), 'Delays'] = "No Delay"

df_reduit.head(5)
```

Ara volem crear una taula amb les aerolínies que acumulen més retards acumulats. Però primer mirarem quantes aerolínies diferents tenim en el nostre dataset.

```python
# Mirem quantes aerolínies tenim al dataset

aerolínies_úniques = df_reduit['UniqueCarrier'].unique()
num_aerolínies = len(aerolínies_úniques)
print(f'Hi ha {num_aerolínies} aerolínies diferents en el conjunt de dades.')
```

```python
# Ara volem saber les 5 primeres aerolínies que han tingut més endarreriments acumulats en l'arribada del vol:

# Agrupem les dades pel nom de l'aerolínia i sumem els endarreriments
grups_aerolínies = df_reduit.groupby('UniqueCarrier')['ArrDelay'].sum()

# Ordenem els resultats en ordre descendent
grups_aerolínies_ordenats = (grups_aerolínies.sort_values(ascending=False) / 60).round(2) # Volem el resultat en hores

# Seleccionem les 5 primeres companyies amb més temps de retard en l'arribada del vol
top_five_airlines = grups_aerolínies_ordenats.head(5)

# Mostrem les 5 primeres companyies amb més temps de retard
print("Les 5 primeres companyies amb més temps de retard acumulat són, en hores:")
print(top_five_airlines)
```

Així doncs, les aerolínies que han patit més temps d'endarreriment acumulat són:

1. WN: Southwest Airlines (EUA <br>
2. AA: American Airlines (EUA) <br>
3. UA: United Airlines (EUA) <br>
4. MQ: American Eagle Airlines (EUA) <br>

```python
# També mirarem les 5 primeres aerolínies que han patit retards més vegades:

# Filtrem les files amb endarreriments
df_endarreriments = df_reduit[df_reduit['Delays'] != 'No delay']

# Comptem les vegades que apareix cada aerolínia 
endarreriments_per_aerolínia = df_endarreriments['UniqueCarrier'].value_counts()

# Ordenem les aerolínies per endarreriments en ordre descendent
aerolínies_mes_endarreriments = endarreriments_per_aerolínia.sort_values(ascending=False)

# Mostrem les tres primeres aerolínies amb més endarreriments
tres_primeres_aerolínies = aerolínies_mes_endarreriments.head(5)

print("Les tres primeres aerolínies amb més endarreriments més vegades són:")
print(tres_primeres_aerolínies)
```

Així doncs, les aerolínies que han patit més endarreriments coincideixen amb les que han patit més temps acumulat d'endarreriment i Southwest Airlines encapçala ambdues llistes.

El que més sorprèn és que Southwest Airlines acumula quasi el doble de retards (en nombre, no en temps) que la següent aerolínia a la llista.

El que volem fer ara és esbrinar la causa principal dels seus retards i per això haurem de fixar-nos en les variables CarrierDelay, WeatherDelay, NASDelay, SecurityDelay i LateAircraftDelay.

```python
# Filtrem les files de la companyia aèria "WN" amb retards
df_wn = df_reduit[(df_reduit['UniqueCarrier'] == 'WN') & (df_reduit['Delays'] != 'No delay')]

# Calculem la suma de cada tipus de retard per aquesta aerolínia en minuts
total_carrier_delay_minuts = df_wn['CarrierDelay'].sum()
total_weather_delay_minuts = df_wn['WeatherDelay'].sum()
total_nas_delay_minuts = df_wn['NASDelay'].sum()
total_security_delay_minuts = df_wn['SecurityDelay'].sum()
total_late_aircraft_delay_minuts = df_wn['LateAircraftDelay'].sum()

# Convertim els valors de minuts a hores dividint per 60
total_carrier_delay_hores = (total_carrier_delay_minuts / 60).round(2)
total_weather_delay_hores = (total_weather_delay_minuts / 60).round(2)
total_nas_delay_hores = (total_nas_delay_minuts / 60).round(2)
total_security_delay_hores = (total_security_delay_minuts / 60).round(2)
total_late_aircraft_delay_hores = (total_late_aircraft_delay_minuts / 60).round(2)

# Mostrem el temps acumulat de retard per cada variable en hores
print("Temps acumulat de retard per a la companyia aèria 'WN' (en hores):")
print(f"CarrierDelay: {total_carrier_delay_hores} hores")
print(f"WeatherDelay: {total_weather_delay_hores} hores")
print(f"NASDelay: {total_nas_delay_hores} hores")
print(f"SecurityDelay: {total_security_delay_hores} hores")
print(f"LateAircraftDelay: {total_late_aircraft_delay_hores} hores")
```

Com podem veure, LateAircraftDelay és la causa prinicipal dels seus retards, amb 11.384 hores. Per tant, l'arribada del vol provinent d'una destinació prèvia provoca més temps de retard en els seus vols. Pensem que això ve donat pel fet que és una companyia àeria de baix cost i necessita minimitzar el temps d'espera entre vols per millorar la seva rentabilitat, ja que el preu del bitllet és més ajustat (oferint més vols diaris, a preu més baix).

Ara volem analitzar quina és la causa principal entre les descrites, que provoca més retards en els vols, com hem fet anteriorment amb els vols de la companyia Continental Airlines.

```python
# Filtrem les files dels vols amb retards
df_retards = df_reduit[(df_reduit['Delays'] != 'No delay')]

# Calculem la suma de cada tipus de retard en minuts
total_carrier_delay_minuts = df_retards['CarrierDelay'].sum()
total_weather_delay_minuts = df_retards['WeatherDelay'].sum()
total_nas_delay_minuts = df_retards['NASDelay'].sum()
total_security_delay_minuts = df_retards['SecurityDelay'].sum()
total_late_aircraft_delay_minuts = df_retards['LateAircraftDelay'].sum()

# Convertim els valors de minuts a hores dividint per 60
total_carrier_delay_hores = (total_carrier_delay_minuts / 60).round(2)
total_weather_delay_hores = (total_weather_delay_minuts / 60).round(2)
total_nas_delay_hores = (total_nas_delay_minuts / 60).round(2)
total_security_delay_hores = (total_security_delay_minuts / 60).round(2)
total_late_aircraft_delay_hores = (total_late_aircraft_delay_minuts / 60).round(2)

# Mostrem el temps acumulat de retard per cada variable en hores
print("Temps acumulat de retard segons les causes (en hores):")
print(f"CarrierDelay: {total_carrier_delay_hores} hores")
print(f"WeatherDelay: {total_weather_delay_hores} hores")
print(f"NASDelay: {total_nas_delay_hores} hores")
print(f"SecurityDelay: {total_security_delay_hores} hores")
print(f"LateAircraftDelay: {total_late_aircraft_delay_hores} hores")
```

Com podem veure, la causa principal dels retards dels vols en general és "late aircraft delay", és a dir l'arribada de l'avió provinent d'una destinació prèvia, en segon lloc "Carrier delay", és a dir un retard per la gestió de l'aerolínia, en tercer lloc "NAS Delay". La següent causa, el temps, acumula moltes menys hores de retard que les anteriors i per últim la causa relacionada amb la seguretat és molt menor.

Podem pensar també que la causa de l'arribada tardia de l'avió es pot deure a "carrier delays" previs, que fan que afecti als vols següents. Ens semblen interessants els resultats obtinguts perquè inicialment podríem pensar que el temps és un dels factors que afecten més en els retards, en canvi és un dels factors més minoritaris.
